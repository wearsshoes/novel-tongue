```python
src_sents = []
tgt_sents = []
alignments = []

with open('c1_word_aligned.txt', 'r') as f:
    lines = f.read().splitlines()
    for i in range(0, len(lines), 4):
        src_sents.append(lines[i])
        tgt_sents.append(lines[i+1])
        alignments.append(ast.literal_eval(lines[i+2]))

print(alignments[0])
```

```python
def recombine_words(src_sentence, tgt_sentence, alignments):
    def process_sentence(sentence):
        words = sentence.split()
        processed_words = []
        for word in words:
            if word.startswith("##"):
                processed_words[-1] += word[2:]
            else:
                processed_words.append(word)
        return processed_words

    src_words = process_sentence(src_sentence)
    tgt_words = process_sentence(tgt_sentence)

    src_mapping = {}
    tgt_mapping = {}
    src_index = 0
    tgt_index = 0

    for i, word in enumerate(src_sentence.split()):
        src_mapping[i] = src_index
        if not word.startswith("##"):
            src_index += 1

    for i, word in enumerate(tgt_sentence.split()):
        tgt_mapping[i] = tgt_index
        if not word.startswith("##"):
            tgt_index += 1

    adjusted_alignments = []
    for src_idx, tgt_idx in alignments:
        adjusted_src_idx = src_mapping[src_idx]
        adjusted_tgt_idx = tgt_mapping[tgt_idx]
        adjusted_alignments.append([adjusted_src_idx, adjusted_tgt_idx])

    return " ".join(src_words), " ".join(tgt_words), adjusted_alignments
```

```python
new_alignments = []

for src_sentence, tgt_sentence, alignment in zip(src_sents, tgt_sents, alignments):
    recombined_src, recombined_tgt, adjusted_alignment = recombine_words(src_sentence, tgt_sentence, alignment)

    new_alignments.append([recombined_src.split(), recombined_tgt.split(), adjusted_alignment])
```

```python
print(new_alignments[99])
print(" ".join(new_alignments[99][1]))

for alignments in new_alignments[99:100]:
    src_tokens = alignments[0]
    tgt_tokens = alignments[1]
    alignment = alignments[2]
    display(show_word_alignments(src_tokens, tgt_tokens, alignment))
```
