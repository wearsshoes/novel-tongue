## Basic example

```python
from bertalign import Bertalign
import torch
print(torch.cuda.is_available())
print(torch.version.cuda)
```

```python
src_file = 'notebooks/alice_in_wonderland/c1_en.txt'
with open(src_file, 'r') as file:
    src = file.read()
```

```python
tgt = 'notebooks/alice_in_wonderland/c1_es.txt'
```

```python
aligner = Bertalign(src, tgt)
aligner.align_sents()
```

```python
src_lang = 'en'
tgt_lang = 'es'
dest_dir = 'notebooks/aligned_texts'

chapters = ["c{}".format(i) for i in range(1, 2)]
for chapter in chapters:
  with open(f"{dest_dir}/{chapter}_{src_lang}.txt", "r") as src_file:
    src = src_file.read()
  with open(f"{chapter}_{tgt_lang}.txt", "r") as tgt_file:
    tgt = tgt_file.read()
  aligner = Bertalign(src, tgt)
  aligner.align_sents()

  aligner_result = ""
  for bead in aligner.result:
      src_line = aligner._get_line(bead[0], aligner.src_sents)
      tgt_line = aligner._get_line(bead[1], aligner.tgt_sents)
      aligner_result += f"{src_line}\n{tgt_line}\n\n"

  with open(f'aligned_texts/{chapter}_aligned.txt', 'w') as file:
    file.write(aligner_result)
```
